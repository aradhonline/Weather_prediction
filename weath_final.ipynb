```python

from darksky import forecast
from datetime import datetime as dt
from datetime import timedelta
import pandas as pd
import numpy as np
#import seaborn as sns
#%matplotlib inline
import time
from random import randint
def process_date(date):
    date_adj = date.isoformat()
    return date_adj.split('.')[0]
    
def forcast_weather(date):
    #return forecast('2650981ba69017d839b692bfbc254bae',25.2134,55.2823,time = process_date(date)) #DIFC
    return forecast('2650981ba69017d839b692bfbc254bae',25.1807346,55.2763149,time = process_date(date)) #BB03
    #return forecast('2650981ba69017d839b692bfbc254bae',25.132215, 55.420699,time = process_date(date)) #BITS PILANI

today = dt(2019,8,1)
date = []
hour = []
day = []
day1_temp = []
day1_humidity = []
day1_wind = []
day1_dewpt = []


for y in range(0,1):
    print(y)
    Dubai_forcast_day1 = forcast_weather(today)
    for x in range(len(Dubai_forcast_day1.hourly)):
        #try:
            date.append(today)
            hour.append(x)
            day.append(today.weekday())
            try:
                day1_temp.append(Dubai_forcast_day1.hourly[x].temperature)
            except:
                day1_temp.append(0)
            try:
                day1_humidity.append(Dubai_forcast_day1.hourly[x].humidity)
            except:
                day1_humidity.append(0)
            try:
                day1_wind.append(Dubai_forcast_day1.hourly[x].windSpeed)
            except:
                day1_wind.append(0)
            try:
                day1_dewpt.append(Dubai_forcast_day1.hourly[x].dewPoint)
            except:
                day1_dewpt.append(0)
        #except:
            #pass
    today = today + timedelta(days=1)


data = pd.DataFrame(data = day1_temp,columns = ['T_d1'])
data['HUM'] = day1_humidity
data['WIND'] = day1_wind
data['DEW_PT'] = day1_dewpt
data['date'] = date
data['hour'] = hour
data['day'] = day
data['DayofYear'] = data['date'].dt.dayofyear
export_csv = data.to_csv (r'C:\Users\Administrator\Desktop\Empower\BB03 WEATHER\1stAug.csv', index = None, header=True)
#print(data)
```
